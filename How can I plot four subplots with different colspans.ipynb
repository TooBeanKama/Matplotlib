```python
import matplotlib.pyplot as plt
```

# subplot2grid

```python
ax1 = plt.subplot2grid((3,2), (0,0), colspan=1)
ax2 = plt.subplot2grid((3,2), (0,1), colspan=1)
ax3 = plt.subplot2grid((3,2), (1,0), colspan=2)
ax4 = plt.subplot2grid((3,2), (2,0), colspan=2)
```
